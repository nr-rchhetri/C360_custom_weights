```python
# 1. Create the base structure first (48 rows total)
metrics = [
    'Null Values', 'Count', 'Minimum', 'Maximum', 'Mean',
    'Standard Deviation', 'P25', 'Median (P50)', 'P75', 'P95','P99',
    'Zero Values', 'Negative Values'
]

columns_to_analyze = [
    'PCT_P1_CASES', 
    'PCT_P1P2_CASES', 
    'RESOLUTION_RATE', 
    'AVG_CSAT_SCORE',
    'MTHLY_FTTR_50', 
    'churn_risk_score', 
    'sentiment_score',
    'engagement_score',
    'OPEN_TICKETS'
]

# Create base DataFrame with exactly 48 rows
result_df = pd.DataFrame({
    'Metric': metrics * 4,
    'Contract_Status': ['Contract'] * 13 + ['Contract'] * 13 + 
                      ['No Contract'] * 13 + ['No Contract'] * 13,
    'Year_Range': ['2023-2024'] * 13 + ['2024-2025'] * 13 + 
                  ['2023-2024'] * 13 + ['2024-2025'] * 13
})

# 2. Create masks for filtering data
period_1_mask = (df['REPORT_AS_OF_DT'] >= '2023-04-01') & (df['REPORT_AS_OF_DT'] <= '2024-03-31')
period_2_mask = (df['REPORT_AS_OF_DT'] >= '2024-04-01') & (df['REPORT_AS_OF_DT'] <= '2025-03-31')
contract_mask = df['EFFECTIVE_ACR_NEW'] > 0
no_contract_mask = df['EFFECTIVE_ACR_NEW'] == 0

# 3. Calculate metrics for each combination
for year_mask, year_label in [(period_1_mask, '2023-2024'), (period_2_mask, '2024-2025')]:
    for status_mask, status_label in [(contract_mask, 'Contract'), (no_contract_mask, 'No Contract')]:
        filtered_data = df[year_mask & status_mask]
        
        for column in columns_to_analyze:
            mask = (result_df['Year_Range'] == year_label) & (result_df['Contract_Status'] == status_label)
            total_count = len(filtered_data)
            
            if total_count > 0:
                # Calculate metrics
                null_count = filtered_data[column].isnull().sum()
                non_null_count = filtered_data[column].notnull().sum()
                zero_count = (filtered_data[column] == 0).sum()
                negative_count = (filtered_data[column] < 0).sum()
                
                # Update values for each metric
                result_df.loc[mask & (result_df['Metric'] == 'Null Values'), column] = f"{null_count} ({(null_count/total_count*100):.2f}%)"
                result_df.loc[mask & (result_df['Metric'] == 'Count'), column] = non_null_count
                result_df.loc[mask & (result_df['Metric'] == 'Zero Values'), column] = f"{zero_count} ({(zero_count/total_count*100):.2f}%)"
                result_df.loc[mask & (result_df['Metric'] == 'Negative Values'), column] = f"{negative_count} ({(negative_count/total_count*100):.2f}%)"
                
                # Non-null values for numerical calculations
                non_null_data = filtered_data[column].dropna()
                if len(non_null_data) > 0:
                    result_df.loc[mask & (result_df['Metric'] == 'Minimum'), column] = round(non_null_data.min(), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'Maximum'), column] = round(non_null_data.max(), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'Mean'), column] = round(non_null_data.mean(), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'Standard Deviation'), column] = round(non_null_data.std(), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'P25'), column] = round(non_null_data.quantile(0.25), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'Median (P50)'), column] = round(non_null_data.quantile(0.5), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'P75'), column] = round(non_null_data.quantile(0.75), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'P95'), column] = round(non_null_data.quantile(0.95), 2)
                    result_df.loc[mask & (result_df['Metric'] == 'P99'), column] = round(non_null_data.quantile(0.99), 2)

# 4. Sort and clean up
result_df = result_df.sort_values(['Year_Range', 'Contract_Status', 'Metric']).reset_index(drop=True)

# Fill any remaining NaN values with 0
result_df = result_df.fillna(0)

print("Final shape:", result_df.shape)
print("\nFirst few rows:")
print(result_df.head(13))  # Show first complete set of metrics

# Save to CSV
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the relevant columns are present and properly named
columns_to_check = [
    'PCT_P1_CASES', 
    'PCT_P1P2_CASES', 
    'RESOLUTION_RATE', 
    'AVG_CSAT_SCORE',
    'MTHLY_FTTR_50', 
    'churn_risk_score', 
    'sentiment_score',
    'engagement_score',
    'OPEN_TICKETS'
]

# If your period columns are named differently, adjust accordingly
period_col = 'Year_Range'
contract_col = 'Contract_Status'

# Convert columns to numeric if needed
for col in columns_to_check:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

# 1. Grouped bar chart: Mean comparison for each metric
means = merged_df.groupby([contract_col, period_col])[columns_to_check].mean().reset_index()

for col in columns_to_check:
    plt.figure(figsize=(7, 4))
    sns.barplot(
        data=means,
        x=contract_col,
        y=col,
        hue=period_col,
        palette='Set2'
    )
    plt.title(f'Mean {col}: 2023-2024 vs 2024-2025 by Contract Type')
    plt.ylabel(f'Mean {col}')
    plt.xlabel('Contract Type')
    plt.legend(title='Year Range')
    plt.tight_layout()
    plt.show()

# 2. Side-by-side box plots for each metric
for col in columns_to_check:
    plt.figure(figsize=(8, 5))
    sns.boxplot(
        data=merged_df,
        x=contract_col,
        y=col,
        hue=period_col,
        palette='Set2'
    )
    plt.title(f'{col}: Distribution by Contract Type and Year Range')
    plt.ylabel(col)
    plt.xlabel('Contract Type')
    plt.legend(title='Year Range')
    plt.tight_layout()
    plt.show()
```
