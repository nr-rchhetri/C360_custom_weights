```python
score_cols = [
    "FINANCIAL_HEALTH_SCORE_CAL",
    "ADOPTION_SCORE_CAL",
    "ENGAGEMENT_SCORE_CAL",
    "RISK_SCORE_CAL"
]
df['OVERALL_HEALTH_SCORE'] = df[score_cols].mean(axis=1)
```

```python
df['REPORT_MONTH'] = pd.to_datetime(df['REPORT_MONTH'])
df = df[df['REPORT_MONTH'] < pd.to_datetime('2024-12-01')]
df.head()
```

```python
# Calculate time-ordered correlation for each account
correlation_data = []

for account in df['SUBSCRIPTION_ACCOUNT_ID'].unique():
    # Get account data and sort by date
    account_data = df[df['SUBSCRIPTION_ACCOUNT_ID'] == account].sort_values('REPORT_AS_OF_DT')
    
    # Only calculate if we have enough data points
    if len(account_data) >= 3:
        correlation = account_data['OVERALL_HEALTH_SCORE'].corr(account_data['PCT_M_PLUS_6_BCM'])
        
        correlation_data.append({
            'SUBSCRIPTION_ACCOUNT_ID': account,
            'SFDC_ACCOUNT_NAME': account_data['SFDC_ACCOUNT_NAME'].iloc[0],
            'Correlation': correlation,
            'Data_Points': len(account_data),
            'Date_Range': f"{account_data['REPORT_AS_OF_DT'].min()} to {account_data['REPORT_AS_OF_DT'].max()}"
        })

# Create DataFrame and sort by correlation
correlation_df = pd.DataFrame(correlation_data)
correlation_df = correlation_df.sort_values('Correlation', ascending=False)
correlation_df = correlation_df.reset_index(drop=True)

# Round correlation to 3 decimal places
correlation_df['Correlation'] = correlation_df['Correlation'].round(3)

# Display the results with date range information
display(correlation_df)
```
